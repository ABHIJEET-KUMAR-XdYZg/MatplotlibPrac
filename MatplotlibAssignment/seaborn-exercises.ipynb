```python
Ques 5: dataset-- diamonds
    Generate the distribution of prices with respect to color for each cut quality.
    ## output should be as shown below
```

```python
diamond=sns.load_dataset('diamonds')
```

```python
sns.barplot(data=diamond,x='cut',y='price',hue='color',
            hue_order=['E', 'I', 'J', 'H', 'F', 'G', 'D'],
            order=['Ideal','Premium','Good','Very Good','Fair'])
plt.legend(ncol=4)
plt.xlabel('Cut',fontsize=14)
plt.ylabel('Price',fontsize=14)
plt.show()
```

```python
Ques 6: dataset-- diamonds
    obtain the counts of diamonds of each cut quality through bar chart.
    ## output should be as shown below
```

```python
sns.set_style('white')
sns.countplot(data=diamond,x='cut',color='red',
              order=['Ideal','Premium','Good','Very Good','Fair'])
sns.despine(top=True,right=True,left=False,bottom=False)
```

```python
Ques 8: dataset-- tips
    Draw a nested violinplot and split the violins for easier comparison between day and total_bill.
    ## output should be as shown below
```

```python
tips=sns.load_dataset('tips')
```

```python
sns.set_style('whitegrid')
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True,
               inner='quartile',palette=['#AECCE4','#FFFCBB'])
sns.despine(top=True,bottom=True,left=True,right=True)
```

```python
Ques 11: dataset-- iris
    Create a scatter plot between sepal length and sepal width categorised by species.
            ## output should be as shown below
```

```python
iris=sns.load_dataset('iris')
```

```python
sns.set_style('white')
sns.scatterplot(data=iris,x='sepal_length',y='sepal_width',hue='species')
sns.despine(top=True,right=True)
plt.xticks(np.arange(5,9))
plt.legend(loc='center right',bbox_to_anchor=(1.2,0.43),frameon=False)
```

```python
Ques 12: dataset-- iris
    Visualize the kernel density estimate of petal length for each species.
    ## output should be as shown below
```

```python
sns.set_style('white')
for specie in iris['species'].unique():
    sns.kdeplot(data=iris[iris['species']==specie],x='petal_length',
                label=specie)
sns.despine(top=True,right=True)
plt.legend(title='species',loc='center right',bbox_to_anchor=(1.24,0.51),
           frameon=False)
```

```python
Ques 13: dataset-- iris
    Study the relationship between sepal length and sepal width for the iris dataset by creating a joint plot.
    ## output should be as shown below
    
```

```python
sns.jointplot(data=iris,x='sepal_length',y='sepal_width',
              marginal_kws={'color': '#9494c6'})
```
